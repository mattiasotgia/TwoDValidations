# Data with v10_06_00_04p04 version


```python

dataWithYZ = uproot.open('plots_muondata_v10_06_00_04p04_production/yz_simulation_debug_beamdata_addedEnergyHist_with_yz_productionVersion.root')
dataNoYZ = uproot.open('plots_muondata_v10_06_00_04p04_production/yz_simulation_debug_beamdata_addedEnergyHist_no_yz_productionVersion.root')

planes = ['EE', 'EW']
views  = ['I1', 'I2', 'C']

scale_colors = plt.cm.managua(np.linspace(0, 1, len(planes)*len(views)))

fig, axs = plot.subplots(
    width=8.5, height=5.5, nrows=len(planes), ncols=len(views),
    xlabel='IDE energy (MeV)', ylabel='Entries / 10 keV',
    # rtitle=f'1.5 GeV $\\mu$'
)

for i, plane in enumerate(planes):
    for j, view in enumerate(views):

        plane_view = f'{plane}_{view}_Energy'

        idx = i * len(views) + j
        print(idx)
        
        hist = dataWithYZ['merge'][plane_view].to_hist()
        hep.histplot(hist, yerr=False, ax=axs[idx], color=tuple(scale_colors[idx]), label=f'{plane} / {view} data (w/ YZ)')

        hist = dataNoYZ['daq:simpleSC'][plane_view].to_hist()
        hep.histplot(hist, yerr=True, ax=axs[idx], color=tuple(scale_colors[idx]), histtype='errorbar', label=f'{plane} / {view} data (w/o YZ)',
                     markeredgecolor='k', marker='o', markersize=3, elinewidth=2)

axs[0].format(ltitle=r'1.5 GeV $\mu$')
axs[2].format(rtitle=r'icaruscode v10_06_00_04p04')

axs.format(yscale='log', xlim=(0, 1.05))
axs.legend(loc='ur', ncols=1)
fig.savefig('plots_muondata_v10_06_00_04p04_production/muon_1.5GeV_plotBoth_ideEnergy.pdf')
```

```python
dataNoYZ = uproot.open('plots_muondata_v10_07_xx_develop/yz_simulation_debug_muondata_addedEnergyHist_noYZ.root')
data = uproot.open('plots_muondata_v10_07_xx_develop/yz_simulation_debug_muondata_addedEnergyHist.root')

from collections import defaultdict
planeSC_map = defaultdict(list)

for i in range(360):
    if data[f'daq:simpleSC{i}'].keys():
        planeSC_map[data[f'daq:simpleSC{i}'].keys()[-1].split(';')[0]].append(f'daq:simpleSC{i}')



for plane in ['EW']:
    for view in ['I2']:
        fig, axs = plot.subplots(
            width=10.5, height=4.5,
            xlabel='IDE energy (MeV)', ylabel='Entries / 10 keV',
            rtitle=f'{plane} / {view} Data\nicaruscode v10_06_00_07_develop', ltitle=f'1.5 GeV $\\mu$'
        )

        plane_view = f'{plane}_{view}_Energy'
        scale_colors = plt.cm.Blues(np.linspace(0, 1, len(planeSC_map[plane_view])))
        hists = []
        colors = []
        for i, key in enumerate(planeSC_map[plane_view]):
            hist = data[key][plane_view].to_hist()
            hists.append(hist)
            colors.append(tuple(scale_colors[i]))
        hep.histplot(hists, yerr=False, ax=axs[0], color=colors, stack=True)
        hist = dataNoYZ['daq:simpleSC'][plane_view].to_hist()
        hep.histplot(hist, yerr=True, ax=axs[0], color=selectedcolor, histtype='errorbar', label=f'{plane} / {view} data (w/o YZ)',
                     markeredgecolor='k', marker='o', markersize=3, elinewidth=2)
        axs.format(yscale='log', xlim=(0, 5.5))
        axs.format(yscale='log',)
        axs.legend(ncols=4, loc='ur')
        fig.savefig(f'plots_muondata_v10_07_xx_develop/muon_1.5GeV_plot_{plane}_{view}_ideEnergy.pdf', dpi=200)
```

```python
from collections import defaultdict

planeSC_map = defaultdict(list)

data = uproot.open('oldData_v10_07_xx_develop/yzSimulationDebugWriterBeamDataByTPC.root')
for i in range(360):
    planeSC_map[data[f'daq:simpleSC{i}'].keys()[0].split(';')[0]].append(f'daq:simpleSC{i}')

for plane in ['WE']:
    for view in ['I2']:
        fig, axs = plot.subplots(
            [*np.arange(1, 31).reshape(6, 5).T, [*[31]*3, *[32]*3], [*[31]*3, *[32]*3], [*[31]*3, *[32]*3]],
            width=20,
            xlabel='TPC Z / 3 cm', ylabel='TPC Y / 3 cm',
        )

        plane_view = f'{plane}_{view}'
        completeHist = data[planeSC_map[plane_view][0]][plane_view].to_hist()
        superImposedPlane, xedges, yedges = completeHist.to_numpy()
        superImposedPlane = (superImposedPlane > 0).astype(int)
        for i, key in enumerate(planeSC_map[plane_view]):
            hist = data[key][plane_view].to_hist()
            if i > 0: 
                tmp, _, _ = hist.to_numpy()
                tmp = (tmp > 0).astype(int)
                superImposedPlane += tmp
                completeHist += hist
            hep.hist2dplot(hist, cbar=False, ax=axs[i], cmap='cividis', cmin=1e-5)
            axs[i].format(aspect='equal', urtitle=(key if i<15 else None), ultitle=(key if i>=15 else None))
        hep.hist2dplot(completeHist, cbar=False, ax=axs[30], cmap='cividis', cmin=1e-5)
        axs[30].format(aspect='equal', ltitle=f'TPC plane {plane} / view {view}')
        
        Z = superImposedPlane.T
        bounds = np.arange(-0.5, Z.max() + 1.5, 1)
        norm = mpl.colors.BoundaryNorm(bounds, plt.get_cmap('coolwarm').N)
        m = axs[31].pcolormesh(xedges, yedges, Z, cmap='coolwarm', norm=norm)
        axs[31].colorbar(m, ticks=np.arange(Z.max() + 1), loc='ul', frame=False, label='N. of passes')
        axs[31].format(aspect='equal')
        
        fig.savefig(f'TPC_{plane_view}.pdf')
```
